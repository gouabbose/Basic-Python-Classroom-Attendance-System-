# Outlier Removal Using IQR

```python
dataset.describe()
```

```python
# IQR means inter quintile range. Diffrence between 1st and 3rd quintile.
# IOR = Q3-Q1
# Min range = Q1-(1.5*IQR)
# Max range = Q3-(1.5*IQR)
```

```python
dataset.shape
```

```python
Q1 = dataset['Longitude'].quantile(0.25)
Q3 = dataset['Longitude'].quantile(0.75)
IQR = Q3-Q1
min_range = Q1-(1.5*IQR)
max_range = Q3+(1.5*IQR)
min_range , max_range
```

```python
Q1 = dataset['Latitude'].quantile(0.25)
Q3 = dataset['Latitude'].quantile(0.75)
IQR = Q3-Q1
min_range = Q1-(1.5*IQR)
max_range = Q3+(1.5*IQR)
min_range , max_range
```

```python
dataset1 = dataset[(dataset['Longitude'] < max_range)]
dataset1 = dataset[(dataset['Latitude'] < max_range)]
dataset1.shape
```

# Naive Bayes Algorithm
It's a  classification algorithm based on Baye's theorem.

**Naive:** Because it assumes that the occurence of a certain feature is independent of the occurence of other features.

**Bayes:** Because it depends on the principle of bayes' theorem.

p(A/B) = [P(B/A) * P(A)] / P(B)

There are three types of Naive Bayes' Model. Such  as:

**Gussian:** Suitable for features that are continuous and have a normal distribution.

**Multinomial:** Assumes that features follow a multinomial distribution. Typically used for discrete data such as text data.

**Bernoulli:** Assumes that features are binary variables. Auaitable for data that can be represented as bianry features suach as document classsification. (One-Hot endcoding gives beat result)


**Conditional Probability:**

*Independent Probability:*  Means the value of the probability of all out comes remain same.

*Dependent Probability:* Means the value of the probability of all out comes remain not same.
